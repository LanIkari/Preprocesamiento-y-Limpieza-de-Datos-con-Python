## SQL

```python
#Importamos pandas
import pandas as pd
```

```python
#Armamos la conexion a la base de datos con todos los atributos
from sqlalchemy import create_engine

us = 'fesalu'
passwd = 'bdfes2023'
url = 'bd.arcelia.net'
puerto = '3306'
bd = 'bdcolegio02' #Cambio de Base de Datos
auth = 'mysql_native_password'

str_conn = 'mysql+mysqlconnector://{0}:{1}@{2}:{3}/{4}?auth_plugin={5}'
str_conn = str_conn.format(us, passwd, url, puerto, bd, auth)
#Comprobamos que nuetra conexion es correcta
str_conn
```

```python
#Creacion del objeto de conexion
conn=create_engine(str_conn)
conn
```

```python
sql = '''SELECT * FROM alumnos'''
df_mysql=pd.read_sql(sql,con=conn)
```

```python
df_mysql
```

```python
sql = '''SELECT * FROM alumnos WHERE sexo = %s '''
df_mysql=pd.read_sql(sql,con=conn, params=['M'])#Obtenemos todos los alumnos que sean hombres
df_alumnos_Hombres=df_mysql #Creacion de un nuevo DATAFRAME
```
